```python
# This is a machine learning project that uses ensemble learning algorithms: random forests, gradient boosting classifier, 
# and XGBoost to predict four different binary classifications in the data: sex, only child status, city vs village, 
# and house vs flat. The accompanying presentation can be viewed at https://youtu.be/2Jwk44KGMXY

# The data is from Kaggle - Young People Survey uploaded by Miroslav Sabo
# https://www.kaggle.com/datasets/miroslavsabo/young-people-survey

# Exploratory data analysis and data cleaning

# Import needed packages

import pandas as pd
import numpy as np
import sklearn

# Set pandas to display all columns; import dataset

pd.set_option('display.max_columns', None)
data = pd.read_csv(r'C:\Users\jejohnson\Documents\Utica\DSC 609\Week 4\archive\responses.csv')

# Look at dataframe as part of EDA

data
```

```python
# Look at all columns vertically as part of EDA

list(data)
```

```python
# Update answer choices under "Lying" column to clarify meaning in preparation for dummy variables and to avoid duplicate 
# column names

data['Lying'].replace('never','never lies',inplace=True)
data['Lying'].replace('sometimes','sometimes lies',inplace=True)
data['Lying'].replace('only to avoid hurting someone','only lies to avoid hurting someone',inplace=True)
data['Lying'].replace('everytime it suits me','lies everytime it suits me',inplace=True)
```

```python
# Update answer choice under "Alcohol" column to clarify meaning in preparation for dummy variables and to avoid duplicate 
# column names

data['Alcohol'].replace('never','never drinks',inplace=True)

# Review dataframe again

data
```

```python
# Get list of columns with data type in preparation for adding dummy variables

column_dtypes = data.columns.to_series().groupby(data.dtypes).groups
column_dtypes
```

```python
# Create dummy variables from all columns with object type

dummy_smoker = pd.get_dummies(data['Smoking'])
dummy_alcohol = pd.get_dummies(data['Alcohol'])
dummy_punctuality = pd.get_dummies(data['Punctuality'])
dummy_lying = pd.get_dummies(data['Lying'])
dummy_internet = pd.get_dummies(data['Internet usage'])
dummy_gender = pd.get_dummies(data['Gender'])
dummy_handed = pd.get_dummies(data['Left - right handed'])
dummy_education = pd.get_dummies(data['Education'])
dummy_onlychild = pd.get_dummies(data['Only child'])
dummy_villagetown = pd.get_dummies(data['Village - town'])
dummy_house = pd.get_dummies(data['House - block of flats'])
```

```python
# Append dummy variables to data

data2 = pd.concat([data, dummy_smoker, dummy_alcohol, dummy_punctuality, dummy_lying, dummy_internet, dummy_handed, dummy_education, dummy_onlychild, dummy_villagetown, dummy_house, dummy_gender], axis=1)
```

```python
# Drop original object-type columns

data3 = data2.drop(['Smoking', 'Alcohol', 'Punctuality', 'Lying', 'Internet usage', 'Left - right handed', 'Education', 'Only child', 'Village - town', 'House - block of flats', 'Gender'], axis=1)

# Review dataframe again

data3
```

```python
# Rename columns "yes" and "no" to specify that they are about the person being an only child or not

data3.rename(columns={'no': 'is not an only child', 'yes': 'is an only child'}, inplace=True)
```

```python
# Review dataframe again with dummy variables

data3
```

```python
# Deal with missing data

# Look at how many rows have at least one missing data point

missing_data_rows = data3.isnull().any(axis = 1).sum()
missing_data_rows_percent = round((missing_data_rows/1010) * 100)

print("Percentage of rows with at least one missing data point: ", missing_data_rows_percent, "%")
```

```python
# Look at how many columns have at least one missing data point

missing_data_cols = data3.isnull().any(axis = 0).sum()
missing_data_cols_percent = round((missing_data_cols/173) * 100)

print("Percentage of columns with at least one missing data point: ", missing_data_cols_percent, "%")
```

```python
# Unfortunately, we have quite a bit of data missing here.

# We could use imputation to fill in the gaps, but because virtually one third of all rows have at least one missing value,
# that would be a lot of guessing. Our best option is probably to eliminate any rows that have at least one missing value. 
```

```python
# Drop all rows that contain at least one NA value and view new dataset (data4)

data4 = data3.dropna(axis=0, how='any')
data4
```

```python
# Female vs Male

# Prep data for ML algorithms to predict sex

# Drop female and male labels from data

X = data4.drop(['female', 'male'], axis=1)
y = data4['female']

# Only either female or male is needed for y, not both, since they are binary (same for other classifications)

# Split data into 70% training, 30% testing

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3) # 70% training and 30% test

# Import scikit-learn metrics module for accuracy calculation

from sklearn import metrics
```

```python
# Female vs Male

# Implementing models

# Random forest with 10, 20, 40, 60, 80, 100, and 120 trees

# Import random forest model

from sklearn.ensemble import RandomForestClassifier

# Run loop with different numbers of trees

est_list = [10, 20, 40, 60, 80, 100, 120]

for estimator in est_list:
    clf=RandomForestClassifier(n_estimators=estimator)
    X_test.fillna(X_train.mean(), inplace=True)
    clf.fit(X_train,y_train)

    print("Number of trees: ", estimator)
    print("Accuracy score (training): {0:.3f}".format(clf.score(X_train, y_train)))
    print("Accuracy score (validation): {0:.3f}".format(clf.score(X_test, y_test)))
    
# Highest accuracy score: Tie between random forest with 80 and 100 trees (both 93%)
```

```python
# Female vs Male

# Ensemble 2: Gradient boosted decision trees

# Import gradient boosting model

from sklearn.ensemble import GradientBoostingClassifier

# Define gradient boosting classifier; experiment with different learning rates (LRs) of GBC to see if we can get the accuracy to increase

lr_list = [0.05, 0.075, 0.1, 0.25, 0.5, 0.75, 1]

for learning_rate in lr_list:
    gb_clf = GradientBoostingClassifier(n_estimators=500, learning_rate=learning_rate, max_features=2, max_depth=2, random_state=0)
    gb_clf.fit(X_train, y_train)

    print("Learning rate: ", learning_rate)
    print("Accuracy score (training): {0:.3f}".format(gb_clf.score(X_train, y_train)))
    print("Accuracy score (validation): {0:.3f}".format(gb_clf.score(X_test, y_test)))
    
# LR of .5 is most accurate (92%)
```

```python
# Female vs Male

# Ensemble 3: XGBoost Classifier

from xgboost import XGBClassifier

xgb_clf = XGBClassifier()
xgb_clf.fit(X_train, y_train)

y_pred=xgb_clf.predict(X_test)

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))

# XGBoost: 92% accurate
```

```python
# Female vs male accuracy winner: random forests with 80 and 100 trees (both 93%)
```
